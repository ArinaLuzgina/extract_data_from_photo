```python
from image_processing_func import process_one_image
import numpy as np
import csv
```

```python
import cv2

img = cv2.imread(".\\images\\image.png")               # загружаем
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # переводим в оттенки серого

# Порог: всё темнее 128 → чёрное, светлее → белое
_, bw = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)

cv2.imwrite(".\\images\\imagebw.png", bw)
```

```python
lower_black = np.array([0, 0, 0])
upper_black = np.array([180, 255, 30])
limits = [[350, 0], [800, 0], [350, 0], [350, 100]]
image_name = ".\\images\\imagebw.png"
process_one_image(
    file_name= image_name,
    save_data=True,
    save_f_name=".\\data\\image_data.csv",
    show_data=True,
    limits=limits,   # axis ranges in physical units
    force_recalibration=False,     # force you to click axis calibration points again
    mask_limits=[lower_black, upper_black],
    interactive_curve=True,
    apply_smoothing=True,       # <-- enable smoothing
    smoothing_bins=900,         # adjust resolution of smoothing. The more the number of bins the longer are calculations but the more they are precise.
    smoothing_method="median",
    morph_kernel_size=1 
)
```
