```python
data_1 = []
with open('.\\data\\elipticity_data.csv', mode ='r')as file:
  csvFile = csv.reader(file, delimiter=";")
  flag = False
  counter = 60
  index = 0
  for lines in csvFile:
        if flag:
          try:
            data_1.append([float(lines[0]), float(lines[1])])
          except Exception as e:
            print(e)
          index += 1
        if index == counter:
           break
        flag = True
data_1 = np.array(data_1)
```

```python
data_2 = []
with open('.\\data\\rotation_data.csv', mode ='r')as file:
  csvFile = csv.reader(file, delimiter=";")
  flag = False
  counter = 60
  index = 0
  for lines in csvFile:
        if flag:
          try:
            data_2.append([float(lines[0]), float(lines[1])])
          except Exception as e:
            print(e)
          index += 1
        if index == counter:
           break
        flag = True
data_2 = np.array(data_2)
```

```python
def find_optimal(data1, w1, data2, w2):
    index = np.argmax(np.abs(data1[:, 1] * w1 + np.abs(data2[:, 1]) * w2))
    return (data1[index], data2[index])
```

```python
print(find_optimal(data_1, 1, data_2, 0))
print(find_optimal(data_1, 0, data_2, 1))
print(find_optimal(data_1, 0.5, data_2, 0.5))
```
